```python
emotions, counts = np.unique(df["Emotion"], return_counts=True)

plt.figure(figsize=(10, 6))
labels = plt.bar(x=emotions, height=counts)
plt.xlabel("Emotion")
plt.ylabel("Number of occurances in dataset")
plt.bar_label(labels)
plt.show()
```
