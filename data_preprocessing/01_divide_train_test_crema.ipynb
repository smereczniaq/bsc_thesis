```python
import os

import shutil
import numpy as np
import pandas as pd
from random import randrange

from sklearn.model_selection import train_test_split
```

```python
source_folder = "../data/kaggle/input/speech-emotion-recognition-en/Crema/"
destination_folder = "../data/divided/Crema/train"

file_names = []
classes = []

df = pd.DataFrame()

for file_name in os.listdir(source_folder):
    # construct full file path
    if file_name.endswith(".wav"):
        file_class = file_name.split('_')[2]
        source = source_folder + file_name
        file_names.append(file_name)
        
        if file_class == 'ANG':
            class_name = "Anger"
        elif file_class == 'DIS':
            class_name = "Disgust"
        elif file_class == 'FEA':
            class_name = "Fear"
        elif file_class == 'HAP':
            class_name = "Happy"
        elif file_class == 'NEU':
            class_name = "Neutral"
        elif file_class == 'SAD':
            class_name = "Sad"

        classes.append(class_name)
        destination = f"{destination_folder}/{class_name}"

        shutil.copy2(source, destination)
```
