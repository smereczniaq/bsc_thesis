# Wczytanie bibliotek

```python
import pandas as pd
import os
import numpy as np
```

# Wczytanie danych

```python
results_path = "../results/"
```

```python
dfs = []

for filename in os.listdir(results_path):
    if ".csv" in filename:
        filepath = results_path + filename
        df = pd.read_csv(filepath)
        nn_type = filename.split("_")[0]
        if nn_type == "no":
            nn_type = "no_dropout"
        optimizer = filename.split("_")[-1].split(".")[0]
        df["nn_type"] = nn_type
        df["optimizer"] = optimizer

        dfs.append(df)
```

```python
results = pd.concat(dfs, ignore_index=True)
results
```

```python
def highlight_max(s): 
    if s.dtype == np.object: 
        is_max = [False for _ in range(s.shape[0])] 
    else: 
        is_max = s == s.max() 
    return ['background: lightgreen' if cell else '' for cell in is_max] 
```

```python
results.style.highlight_max(color='lightgreen', axis=0, subset=["Sum", "Crema", "Ravdess", "Savee", "Tess"])
```
